```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
class PemusatanData:
  class Raw:
    def __init__(self,data):
      get_data = data.split(',')
      self.raw = []
      for i in get_data:
        self.raw.append(int(i))
      self.raw.sort()
      panjang_data = len(self.raw)
      sum_data = 0
      for i in self.raw:
        sum_data = sum_data + i

      #Mean
      self.me = sum_data / panjang_data
      
      #Median
      if panjang_data % 2 == 0:
        t = panjang_data / 2
        t = int(t)
        self.med1 = self.raw[t-1]
        self.med2 = self.raw[t]
        self.med = (self.med1 + self.med2) / 2
      else:
        self.med = self.raw[int(panjang_data/2)]
     
      #Modus
      f = []
      for i in self.raw:
        f.append(self.raw.count(i))
      fmax = max(f)
      if f.count(fmax) == len(f):
        self.mo = "Tidak terdefinisi"
      else:
        fbil = []
        idx = []
        for i in range(len(f)):
          if f[i] == fmax:
            fbil.append(self.raw[i])
            idx.append(i)
        dominan = [self.raw[idx[0]]]
        for i in range(len(idx)-1):
          if self.raw[idx[i]] != self.raw[idx[i+1]]:
            dominan.append(self.raw[idx[i+1]])
        self.mo = dominan
      print("Mean : ",self.me,"\nMedian : ",self.med,"\nModus : ",self.mo)
  

  class Tabel:
    class Tunggal:
      def __init__(self):
        print("Input nilai x :")
        get_x = input()
        x_split = get_x.split(',')
        self.x = []
        for i in x_split:
          self.x.append(int(i))

        print("Input frekuensi (banyak frekuensi harus sama dengan banyak nilai x) : ")
        get_f = input()
        f_split = get_f.split(',')
        if len(f_split) == len(self.x):
          self.f = []
          for i in f_split:
            self.f.append(int(i))
          
          self.fx = []
          for i in range(len(self.x)):
            self.fx.append(self.x[i] * self.f[i])
          
          self.sigmaF = 0
          for i in self.f:
            self.sigmaF = self.sigmaF + i

          self.sigmaFX = 0
          for i in self.fx:
            self.sigmaFX = self.sigmaFX + i

          #Mean
          self.me = self.sigmaFX / self.sigmaF

          #Median
          self.raw = []
          for i in range(len(self.x)):
            for j in range(self.f[i]):
              self.raw.append(self.x[i])
          n = self.sigmaF
          med1 = n / 2
          med2 = med1 + 1
          self.med = int((med1 + med2) / 2)
          self.medValue = self.raw[self.med]
          
          #Modus
          moFrekuensi = max(self.f)
          moIdx = 0
          for i in range(len(self.f)):
            if self.f[i] == moFrekuensi:
              moIdx = i
          self.mo = self.x[moIdx]
          
          #Output
          print("Fx : ",self.fx)
          print("Ftotal : ",self.sigmaF,"\nFXtotal : ",self.sigmaFX)
          print("Mean : ",self.me,"\nMedian : ",self.medValue," pada index ke - ",self.med,"\nModus : ",self.mo," dengan jumlah frekuensi ",self.f[moIdx])
        else:
          print("banyak nilai x harus sama dengan banyak frekuensi")
    
    class Kelompok:
      def __init__(self):
        print("Input batas bawah")
        get_bb = input()
        bb_split = get_bb.split(',')
        self.bb = []
        for i in bb_split:
          self.bb.append(int(i))
        print("Input batas atas")
        get_ba = input()
        ba_split = get_ba.split(',')
        self.ba = []
        for i in ba_split:
          self.ba.append(int(i))
        print("Input Frekuensi")
        get_f = input()
        f_split = get_f.split(',')
        self.f = []
        for i in f_split:
          self.f.append(int(i))
        
        if len(self.bb) == len(self.ba) and len(self.ba) == len(self.f):
          #Nilai tengah
          self.midValue = []
          for i in range(len(self.bb)):
            self.midValue.append((self.bb[i] + self.ba[i]) / 2)

          #Raw data
          self.raw = []
          for i in range(len(self.midValue)):
            for j in range(self.f[i]):
              self.raw.append(int(self.midValue[i]))
          
          #FX
          self.fx = []
          for i in range(len(self.midValue)):
            self.fx.append(self.midValue[i] * self.f[i])

          #F total
          self.sigmaF = 0
          for i in self.f:
            self.sigmaF = self.sigmaF + i
      
          #Fx total
          self.sigmaFX = 0
          for i in self.fx:
            self.sigmaFX = self.sigmaFX + i

          #Mean
          self.me = self.sigmaFX / self.sigmaF

          #Median
          medIdx = self.sigmaF / 2
          self.med = self.raw[int(medIdx)]
          
          #Modus
          fMax = max(self.f)
          moIdx = 0
          for i in range(len(self.f)):
            if self.f[i] == fMax:
              moIdx = i
          self.mo = self.midValue[moIdx]
          
          #Output
          print("Ftotal: ",self.sigmaF,"\nFx: ",self.fx,"\nFXtotal: ",self.sigmaFX)
          print("Mean: ",self.me,"\nMedian: ",self.med,"\nModus: ",self.mo," dengan frekuensi ",fMax)
        else:
          print("eror")


print("•"*30)
print("Pilih metode utk memasukan data (pilih 1/2/3)")
print("1.Data deret\n2.Data tabel tunggal\n3.Data tabel kelompok")
choose = int(input())
if choose == 1:
  print("Input data deret (ex: 99,99,99,..)")
  data_user = input()
  PemusatanData.Raw(data_user)
elif choose == 2:
  PemusatanData.Tabel.Tunggal()
elif choose == 3:
  PemusatanData.Tabel.Kelompok()
```
