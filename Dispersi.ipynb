```python
class Dispersi:
  class Persentil:
    class Tunggal:
      def __init__(self):
        print("Input data")
        getData = input()
        dataSplit = getData.split(',')
        print("Berapa banyak Persentil yang dibutuhkan?")
        np = int(input())
        ps = []
        for i in range(np):
          print("Input persentil pada..")
          getPs = int(input())
          ps.append(getPs)
        #Output tampilkan Persentil yang diberikan
        print("Persentil yang diberikan :")
        for i in range(len(ps)):
          print(" ~ P-{}".format(ps[i]))

        #Raw
        self.raw = []
        for i in dataSplit:
          self.raw.append(int(i))
        #sort Raw
        self.rawSort = self.raw
        self.rawSort.sort()
 
        #Panjang data / n
        self.n = len(self.raw)
         
        #Output
        self.persentil = []
        self.urutan = []
        self.jalan = []
        for i in range(len(ps)):
          idx = ps[i] * (self.n + 1) / 100
          self.urutan.append(idx)
          j1 = "{}({}+1) / 100".format(ps[i],self.n)
          j2 = "({}+{}) / 100".format(ps[i] * self.n,ps[i] * 1)
          j3 = "{} / 100".format((ps[i] * self.n) + (ps[i] * 1))
          hasil1 = "= {}".format(idx)
          letak = "Terletak pada urutan ke - {}".format(int(idx))
          ############################
          inf1 = "Dimana data pada urutan ke {}".format(self.urutan[i])
          self.persentil.append(self.rawSort[int(self.urutan[i])-1] + (self.urutan[i] - int(self.urutan[i])))
          j4 = "Yaitu {}".format(self.rawSort[int(self.urutan[i])-1])
          self.jalan.append("{}\n{}\n{}\n{}\n{}\n{}\nYaitu {}".format(j1,j2,j3,hasil1,letak,inf1,self.persentil[i]))
        
        #Range
        xmax = max(self.rawSort)
        xmin = min(self.rawSort)
        self.r = xmax - xmin
        #Output range
        print("Data dengan range {}".format(self.r))
       
        print("n = {}".format(self.n))

        #Output jalan
        for i in range(len(self.jalan)):
          print("•"*20)
          print("\n")
          print("Persentil pada {} adalah sebagai berikut...".format(ps[i]))
          print(self.jalan[i])
          print("\n")
          print("•"*20)

    class Kelompok:
      def __init__(self):
        #Batas bawah
        print("Input batas bawah")
        get_bb = input()
        bbSplit = get_bb.split(',')
        self.bb = []
        for i in bbSplit:
          self.bb.append(int(i))
        
        #Batas atas
        print("Input batas atas")
        get_ba = input()
        baSplit = get_ba.split(',')
        self.ba = []
        for i in baSplit:
          self.ba.append(int(i))

        #Frekuensi
        print("Input frekuensi")
        get_f = input()
        fSplit = get_f.split(',')
        self.f = []
        for i in fSplit:
          self.f.append(int(i))

        if len(self.bb) != len(self.ba) and (len(self.bb) + len(self.ba)) / 2 != len(self.f):
          print("Panjang data yang anda masukan tidak sama!")
        else:
          sumF = 0
          for i in self.f:
            sumF = sumF + i

          #Cetak tabel
          print("._____________.____________.___________.")
          print("                                       .")
          print("| Batas Bawah | Batas Atas | Frekuensi |")
          print("._____________.____________.___________.")
          for i in range(len(self.f)):
            print("|     {}       |     {}      |     {}     |".format(self.bb[i], self.ba[i], self.f[i]))
          print("._____________.____________.___________.")
          print("|   Jumlah    |            |     {}    |".format(sumF))
          print("._____________.____________.___________.")

          #Raw dengan acuan nilai tengah
          mid = []
          for i in range(len(self.f)):
            mid.append((self.bb[i] + self.ba[i]) / 2)
          raw = []
          for i in range(len(mid)):
            for j in range(int(self.f[int(i)])):
              raw.append(mid[i])

          print("Input banyak Persentil yang dibutuhkan")
          get_np = int(input())
          np = []
          for i in range(get_np):
            print("Input Persentil")
            getp = int(input())
            np.append(getp)
          print("Persentil yang diberikan :")
          for i in range(len(np)):
            print(" ~P{}".format(np[i]))

          for i in range(len(np)):
            self.idx = ((np[i] * sumF) + (np[i] * 1)) / 100
            print("Letak P{} => {}({}+1) / 100".format(np[i], np[i], sumF))
            print("             ({}+{}) / 100".format(np[i]*sumF, np[i]*1))
            print("             {} / 100".format((np[i]*sumF) + (np[i]*1)))
            print("             {}".format(self.idx))
            idx2 = raw[int(self.idx)-1]
            letakKelas = 0
            for j in range(len(self.f)):
              if self.bb[j] < idx2 and self.ba[j] > idx2:
                letakKelas = j
            b = self.bb[letakKelas] - 0.5
            fbefore = int(self.f[letakKelas - 1])
            fkelas = int(self.f[letakKelas])
            pkelas = self.bb[1] - self.bb[0]
            print("Letak kelas P{} Terletak pada {} - {}".format(np[i], self.bb[letakKelas], self.ba[letakKelas]))
            print("b / tepi bawah kelas P{} = {}".format(np[i], b))
            print("F / Frekuensi sebelum P{} = {}".format(np[i], fbefore))
            print("f / Frekuensi pada P{} = {}".format(np[i], fkelas))
            print("P / Panjang kelas = {}".format(pkelas))
            print("P{} = {} + {} [({}/100){} - {} / {}]".format(np[i], b, int(pkelas), int(np[i]), int(sumF), int(fbefore), int(fkelas)))
            print("      {} + {} [({} - {}) / {}]".format(b, int(pkelas), (np[i] / 100) * sumF, int(fbefore), int(fkelas)))
            print("      {} + {} [{}]".format(b, int(pkelas), (((np[i]/100)*sumF) - fbefore) / fkelas))
            print("      {} + {}".format(b, pkelas * ((((np[i]/100)*sumF)-fbefore)/fkelas)))
            kurungup = ((np[i]/100) * sumF) - fbefore
            ps = b + (pkelas * (kurungup/fkelas))
            print("P{} = {}".format(np[i], ps))

  class SR:
    class Tunggal:
      def __init__(self):
        print("Input data")
        getData = input()
        dataSplit = getData.split(',')
        data = []
        for i in dataSplit:
          data.append(int(i))
        data.sort()
        
        #xm
        sumData = 0
        for i in data:
          sumData = sumData + i
        lenData = len(data)
        xm = sumData / lenData

        #SR
        sigmax = []
        for i in range(len(data)):
          sigmax.append(data[i] - xm)
          #Mengubah semua bilangan yang negatif menjadi positif
          if sigmax[i] < 0:
            sigmax[i] = sigmax[i] * (-1)
       
        sumx = 0
        for i in range(len(sigmax)):
          sumx = sumx + sigmax[i]
        
        sr = sumx / lenData
        
        #Cetak jalan
        #xm
        #jalan 1
        print("Xm = (", end="")
        for i in range(len(data)-1):
          print("{}".format(data[i]), end="+")
        print("{}".format(data[len(data)-1]), end=")")
        print(" / {}".format(lenData))
        #jalan 2
        print("     {} / {}".format(sumData, lenData))
        print("     {}".format(xm))
       
        #SR
        #jalan 1
        print("SR = ", end="")
        print("[", end="")
        for i in range(len(data)-1):
          print("({} - {}) + ".format(data[i], xm), end="")
        print("({} - {})] / {}".format(data[lenData-1], xm, lenData))
        #jalan 2
        print("   = ", end="(")
        for i in range(len(sigmax)-1):
          print("{} +".format(sigmax[i]), end="")
        print("{}) ".format(sigmax[len(sigmax)-1]), end="")
        print("/ {}".format(lenData))
        #jalan 3
        print("   = {} / {}".format(sumx, lenData))
        print("   = {}".format(sr))

    class Kelompok:
      def __init__(self):
        print("Input batas bawah")
        getbb = input()
        bbsplit = getbb.split(',')
        bb = []
        for i in bbsplit:
          bb.append(int(i))
        print("Input batas atas")
        getba = input()
        basplit = getba.split(',')
        ba = []
        for i in basplit:
          ba.append(int(i))
        print("Input f")
        getf = input()
        fsplit = getf.split(',')

        #f
        f = []
        for i in fsplit:
          f.append(int(i))
        ftotal = 0
        for i in f:
          ftotal = ftotal + i
        
        #x
        x = []
        for i in range(len(f)):
          x.append((bb[i] + ba[i]) / 2)

        #x.f
        x_f = []
        for i in range(len(x)):
          x_f.append(f[i]*x[i])
        xftotal = 0
        for i in x_f:
          xftotal = xftotal + i

        #xmedian
        xm = xftotal / ftotal

        #Cetak jalan xmedian
        print("xMedian -> £x.f / £f")
        print("        -> {} / {}".format(xftotal, ftotal))
        print("        -> {}".format(xm))

        #|x-xm|
        xminxm = []
        for i in range(len(x)):
          xminxm.append(x[i] - xm)
          if xminxm[i] < 0:
            xminxm[i] = xminxm[i] * (-1)
        #x-xm2
        xminxm2 = []
        for i in xminxm:
          xminxm2.append(i**2)

        #f|x-xm|
        fxminxm = []
        for i in range(len(xminxm)):
          fxminxm.append(f[i] * xminxm[i])
        fxminxmtotal = 0
        for i in fxminxm:
          fxminxmtotal = fxminxmtotal + i

        #f|x-xm|°2
        fxminxm2 = []
        for i in range(len(xminxm2)):
          fxminxm2.append(f[i]*xminxm2[i])
        fxminxm2total = 0
        for i in fxminxm2:
          fxminxm2total = fxminxm2total + i

        #Sr
        sr = fxminxmtotal / ftotal
        #Sd
        sd = math.sqrt(fxminxm2total / ftotal)

        #Cetak jalan SR
        print("SR -> €f|x-xm| / €f")
        print("   -> {} / {}".format(fxminxmtotal, ftotal))
        print("   -> {}".format(sr))

        #Cetak jalan Sd
        print("SD -> √(€f|x-xm|^2) / €f")
        print("   -> √{} / {}".format(fxminxm2total, ftotal))
        print("   -> √{}".format(fxminxm2total/ftotal))
        print("   -> {}".format(sd))
    

        #buat table
        tabel = {
          'bb':bb,
          'ba':ba,
          'x':x,
          'f':f,
          'x*f':x_f,
          '|x-xm|':xminxm,
          '|x-xm|2':xminxm2,
          'f|x-xm|':fxminxm,
          'f|x-xm|2': fxminxm2
        }
        for i in tabel:
          if i == 'bb':
            tabel[i].append('Total')
          elif i == 'f':
            tabel[i].append(ftotal)
          elif i == 'x*f':
            tabel[i].append(xftotal)
          elif i == 'f|x-xm|':
            tabel[i].append(fxminxmtotal)
          elif i == 'f|x-xm|2':
            tabel[i].append(fxminxm2total)
          else:
            tabel[i].append("")

        #Show table
      
        self.show = pd.DataFrame(tabel)
```
